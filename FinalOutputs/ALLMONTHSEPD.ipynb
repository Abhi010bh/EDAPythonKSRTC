```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Monthly.csv")
data
```

```python
data.columns=['Months','Count','Mean','Standard Deviation','Minimum','Maximum','Coefficient of Correlation']
data
```

```python
data.index=data['Months']
data
```

```python
data.drop(columns='Months',inplace=True)
data
Order=['Sept2021','Oct2021','Nov2021','Dec2021','JAN2022','Feb2022','April2022','May2022','June2022']
data=data.reindex(Order)
data['MonthID'] = range(1, len(data) + 1)
```

```python

data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X=data['MonthID']
Y=data['Mean']
sns.barplot(x=X,y=Y,orient='v')
plt.xticks(X)
plt.ylim(0, 500)  # Set the y-axis range from 0 to 100
plt.yticks(range(0, 500, 25))
plt.show()
```

```python
print('Mean of Effective Passenger Density',data['Mean'].mean())
```

```python
X=data['MonthID']
Y=data['Coefficient of Correlation']
sns.barplot(x=X,y=Y,orient='v')
plt.xticks(X)
plt.ylim(0, 1)  # Set the y-axis range from 0 to 100
plt.yticks(np.arange(0, 1,0.05))
plt.show()
```

```python

plt.scatter(data['MonthID'], data['Mean'], color='blue', alpha=0.5,label='Mean')
plt.scatter(data['MonthID'], data['Standard Deviation'], color='red', alpha=0.5,label='Standard Deviations')
plt.xlabel('MonthID')
plt.title('Means and Standard Deviations trend')
plt.grid(True)
plt.legend()
plt.show()
```
